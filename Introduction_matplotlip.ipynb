# 1 Pyplot API

```python
x = np.linspace(0, 10, 1000) # tạo array x chạy từ 0 đến 10 và là 1000 số 
plt.plot(x, np.sin(x), color="blue", linestyle="dashed", label="sin(x)");
#vẽ đồ thị trục ngang là x, trục dọc là sin(x), label="sin(x)" : để thêm ghi chú
plt.plot(x, np.cos(x), color="red", label="cos(x)")
plt.title("Sin(X) and Cos(X)") # tên của đồ thị
plt.xlabel("X");# tên của trục ngang là x
plt.ylabel("F(x)"); # tên của trục dọc là F(x)
# những câu lệnh điều chỉnh độ dài 2 trục ngang và dọc
#plt.xlim([0,4])
#plt.ylim([-0.7, 0.6])
#plt.axis([0, 4, -0.7, 0.6]) # [xmin,xmax,ymin,ymax]
#plt.axis("tight") # cho hai trục khít nhau hơn
plt.axis("equal") # độ dài hai trục bằng nhau


plt.legend(); # câu lệnh để hiển thị label 
```

```python
# OO API
# prepare new data
rng = np.random.RandomState() # tạo giá trị x, y ngẫn nhiên
x = rng.randn(100) # cho x ra ngẫu nhiên
y = rng.randn(100) # y ra ngẫu nhiên
sizes = 500*rng.randn(100) # cho mỗi cặp x y có kích cỡ khách nhau
colors = rng.randn(100) # tạo những màu khác nhau cho từng cặp xy
fig, ax=plt.subplots()
img1=ax.scatter(x,y, s=sizes, c=colors, cmap='viridis', alpha=0.3); # cmap để thay đổi bảng màu của biểu đồ, alpha chỉnh độ trong suốt
fig.colorbar(img1); # tạo thêm thanh color bar

```

### project: data visualization - california citi

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('./california_cities.csv')
cities.head()
cities['area_total_km2'].sort_values(ascending=False)
```

```python
# Extract ldtd(vĩ độ), longd(kinh độ)

lat, long = cities['latd'], cities['longd']
#lat[:10]
#long[:10]

# plot using pylot API
plt.scatter(long,lat); # plot lên trên bản đồ
```

```python
# Extract ldtd(vĩ độ), longd(kinh độ)
lat, long = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
plt.style.use('seaborn') # backround của biểu đồ
plt.figure(figsize=(8,6))
# plot using pylot API
plt.scatter(long,lat, 
            c=np.log10(population), # dân số lớn nên dùm hàm log để giảm số lượng dân số xuống mới thể hiện trên bản đồ rõ hơn
            cmap='viridis', s=area, linewidths=0, alpha=0.5); # plot lên trên bản đồ, màu sắc thể hiện dân số, kích cỡ thể hiện diện tích area
plt.axis('equal')
plt.xlabel('kinh độ')
plt.ylabel('vĩ độ')
plt.colorbar(label='log$_{10}$(population)') # log của 10, so sánh màu với cột rồi 10^số trên cột khi so với màu
plt.clim(3,7); # scale thanh colorbar 

# tạo legend cho những kích cỡ của thành phố
area_range=[50,100,300,500] # tạo khoảng diện tích 

for area in area_range: # vẽ từng vòng tròn thể hiện diện tích==> chú thích trong biểu đồ
    plt.scatter([],[], s=area, label= str(area) + 'km$^2$', c='r', alpha=0.4)
    plt.legend(labelspacing=1, title='city area') # hiển thị label lên 
    


plt.title('california cities : population and area distribution');
```
